```python
from sklearn.impute import SimpleImputer as si
```

```python
imputer = si(strategy = 'median')
```

```python
cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
```

```python
diabetes[cols] = diabetes[cols].replace(0, np.nan)
```

```python
diabetes.isna().sum()
```

```python
diabetes[cols] = imputer.fit_transform(diabetes[cols])
```

```python
diabetes.isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
num_cols = diabetes.select_dtypes(exclude = 'object').columns.tolist()
```

```python
diabetes[num_cols] == diabetes
```

```python
inputs = diabetes.drop('Outcome', axis = 1)
targets = diabetes.Outcome
scaler.fit(inputs)
```

```python
num_cols.remove('Outcome')
inputs[num_cols] = scaler.transform(inputs[num_cols])
```

```python
inputs
```

```python
inputs.describe()
```

```python
from sklearn.linear_model import LogisticRegression as lr
```

```python
model = lr(class_weight = 'balanced')
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
train_ins, test_ins, train_trgs, test_trgs = tts(inputs, targets, test_size = 0.2, random_state = 42)
```

```python
model.fit(train_ins, train_trgs)
```

```python
model.score(train_ins, train_trgs)
```

```python
model.score(test_ins, test_trgs)
```
